## Data Exploration

```python
sns.set_style("whitegrid")

ax = sns.countplot(x = 'Size', data=df, palette = 'viridis', hue = 'Size', legend = False)

# Checking for data labels
for bars in ax.containers:
    ax.bar_label(bars, fmt = '%d', label_type = 'edge', fontsize = 10, color = 'black')
```

#### Conclusion: Most of the people buys M-Size

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Courier Status', hue = 'Status')

# Checking for data labels
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

#### Majority to the orders are shipped through the courier

```python
from collections import Counter

# Convert category column to string (if not already)
df['Category'] = df['Category'].astype(str)

# Count occurrences of each category
category_counts = Counter(df['Category'])

# Generate a color palette
colors = sns.color_palette("viridis", len(category_counts))

# Plot the bar chart with multiple colors
plt.figure(figsize=(10, 6))
plt.bar(category_counts.keys(), category_counts.values(), color=colors, edgecolor="red")

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

#### Most of the buyers are buys t-shirt

```python
check_B2B = df['B2B'].value_counts()

#create pie chart
plt.pie(check_B2B, labels = check_B2B.index, autopct = '%1.1f%%')
plt.show()
```

#### Maximum 99.2% of buyers are retailers and 0.8% are wholesalers

```python
x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available size')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data = df, x = "ship-state", palette = 'husl', hue = 'ship-state', legend = False)
plt.xlabel('State')
plt.ylabel('Order_Count')
plt.title('Distribution of state')
plt.xticks(rotation=90, size=12)
plt.show()
```

```python
#Find top-10 from upper chart
top10_state = df['ship-state'].value_counts().head(10)     #for top 10 states

plt.figure(figsize=(15,10))
sns.countplot(data = df[df['ship-state'].isin(top10_state.index)], 
              x = "ship-state", 
              palette = 'viridis', 
              hue = 'ship-state', 
              legend = False)
plt.xlabel('State')
plt.ylabel('Order_Count')
plt.title('Distribution of state')
plt.xticks(rotation=90, size=12)
plt.show()
```

#### Most of the Buyers are from Maharastra state
